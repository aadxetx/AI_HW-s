```python
##################### QUESTION - 1 ######################
import numpy as np
data = np.array([[1,0,0,7,'Apple'], [0,1,0,20,'Watermelon'],[1,0,0,1,'Cherry'],[0,1,0,7.5,'Apple'],[1,0,0,1,'Strawberry'],[1,0,0,0.8,'Cherry']])
```

```python
print(data)
```

```python
x = data[:,:4]
y = data[:,4]
```

```python
print(x)
print(y)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(bootstrap = 'true', max_features = 'sqrt', max_depth = 4)
model.fit(x,y)
```

```python
test = np.array([[1,0,0,10]])
print(test)
```

model.predict(test)

```python
model.predict(test)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
estimator = model.estimators_[5]
plt.figure(figsize = (25,10))
a = plot_tree(estimator, feature_names = ['Red', 'Green', 'Blue', 'Size'], 
              class_names = ['Apple', 'Watermelon','Cherry', 'Strawberry'],
              filled = True,
              rounded = True,
              fontsize = 14)
                                                                                      
```

```python
#################### QUESTION - 2 #######################
import numpy as np
import matplotlib.pyplot as plt
import math



y=[] # to store the values of the function
x=np.arange(-2,2,0.05) # stroing the x values in the x - numpy array
for i in range(len(x)):
     x[i]=round(x[i],2)

print(x) # printing the x values obtained in the range [-2,2] with step of 0.05

for i in x:
    y.append(round(math.e -i**2 + 0.01*(math.cos(200*i)),4)) # calculating the y value of the function f(x)
    
print(y) # printing the y value 

print('\nMaximum value of the function is :',end=" ") 
print(max(y)) # maximum value of the function

plt.xlabel("x ---->")
plt.ylabel('Y=f(x)')

plt.scatter(x,y)
plt.plot(x,y,'r')
plt.show()
```
